```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris

sns.set()
```

```python
x = np.linspace(0, 10, 100)
fig, ax = plt.subplots(figsize=(10, 8))
plt.plot(x, np.cos(x), label='cosine')
plt.plot(x, np.sin(x),'--', label='sine')
plt.plot(x, np.tan(x), '-.', label='tan')
plt.legend()
plt.show()
```

#  Simple Line plot

Every typical plot is made-up of a figure() object and an axes object. The figure() object is a container like object that holds everything needed for the sucessful imlementation of a plot, it contains information about the plot, graphics, text and labels.
The axes() object is resposible for the X and Y co-ordiates.

```python
# basic plot
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x, np.sin(x))
plt.show()
```

Another way of plotting is using the plot() object

```python
plt.plot(x, np.sin(x))
plt.show()
```

### Adjusting the plot: Line Color and Style

```python
# changing the color of the plots with the keyword : color
plt.plot(x, np.sin(x-0), color='blue')
plt.plot(x, np.sin(x-1), color='g')
plt.plot(x, np.sin(x-2), color='0.75')
plt.plot(x, np.sin(x-3), color='#FFDD44')
plt.plot(x, np.sin(x-4), color=(1.0, 0.2, 0.3))
plt.plot(x, np.sin(x-5), color='chartreuse')
plt.show()
```

```python
# changing the line style with the keyword : linestyle
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')
plt.show()
```

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
plt.show()
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5,  1.5])
```

```python
# putting all together
plt.plot(x, np.sin(x), '-', color='green', label='sin(x)')
plt.plot(x, np.cos(x), '--b', label='cos(x)')
plt.axis('equal')

plt.legend()
plt.show()
```

###  Labeling plot

```python
plt.plot(x, np.sin(x), '--b', label='sin(X)')
plt.title('Plot of Sin(X)')
plt.xlabel('X')
plt.ylabel('sin(x)')

plt.legend()
plt.show()
```

```python
ax = plt.axes()
ax.set(title='Simple Plot', xlabel='X', 
      ylabel='Sin X')
ax.plot(x, np.sin(x))

plt.show()
```

###  Scatter plots

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'pk')
plt.title('Simple Scatter plot')
plt.show()
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label='marker={}'.format(marker))
plt.xlim([-.2, 1.8])
plt.legend(numpoints=1)
plt.show()
```

```python
plt.plot(x, y, '-p', color='gray',
         markersize=15,
         markerfacecolor='white',
         markeredgecolor='gray',
        markeredgewidth=2.0)
plt.show()
```

```python
# using plt.scatter to create a scatter plot
plt.scatter(x, y, marker='o', color='k')
plt.show()
```

```python
# plot's size, color and transparency level
x = rng.rand(100)
y = rng.rand(100)
colors = rng.rand(100)
size = 1000 * rng.rand(100)
plt.scatter(x, y, c=colors, s=size, alpha=0.3,
            cmap='jet')
plt.colorbar()
plt.show()
```

```python
iris = load_iris()
features = iris['data'].T
plt.scatter(features[0], features[1], c=iris['target'], 
            s=100*features[3], cmap='jet', alpha=0.5)
plt.xlabel(iris['feature_names'][0])
plt.ylabel(iris['feature_names'][1])
plt.colorbar()
plt.show()
```

### Histograms, Binning, and Density

```python
data = np.random.randn(1000)
plt.hist(data, bins=10, density=False)
plt.show()
```

```python
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)

plt.plot(x, y)
plt.show()
```
