# 1. Problem Definition 
Predicting the price of second-hand Fiat cars using machine learning.

# 2. Data :
The data is gotten from the [kaggle](https://www.kaggle.com/paolocons/another-fiat-500-dataset-1538-rows) website.

# 3. Performance Measurement:
The data would be measured used Root Mean Square Error (RMSE) and a accuracy of about 95% is what I'm trying to attain.

# 4. Features:
The data set is made up 8 columns of  different features of the car and a total of 1538 car samples. According to the repository the dataset has the following description.


>**`Description of colums`**
    1. model: Fiat 500 comes in several 'flavours' :'pop', 'lounge', 'sport'    
    2. engine_power: number of Kw of the engine    
    3. ageindays: age of the car in number of days (from the time the dataset has been
    reated)    
    4. km: kilometers of the car
    5. previous_owners: number of previous owners
    6 lat: latitude of the seller (the price of cars in Italy varies from North to South of the country)
    7 lon: longitude of the seller (the price of cars in Italy varies from North to South of the country)
    8 price: selling price (the target)

```python
def load_data():
    filepath = '../data/fiat car/'
    filename = 'automobile_fiat_dataset.csv'
    file = os.path.join(filepath, filename)
    return pd.read_csv(file, low_memory=True, parse_dates=True, sep=',')
data = load_data()
```

# Building a model 
Linear Models

```python
train = train_set.drop('price', axis=1)
label = train_set['price'].copy()
```

```python
# Ridge Regression
num_attr = [column for column in train_set.columns if not train_set[column].dtype == 'object']
cat_attr = [column for column in train_set.columns if train_set[column].dtype == 'object']

num_processing = Pipeline([
    ('num', SimpleImputer(strategy='mean')),
    ('poly', PolynomialFeatures(degree=3, include_bias=False)),
    ('scale', StandardScaler()),
])

cat_processing = Pipeline([
    ('onehot', OneHotEncoder())
])

data_processing = ColumnTransformer([
    ('num_feature', num_processing, num_attr),
    ('cat_features', cat_processing, cat_attr), 
])

train_prepared = data_processing.fit_transform(train_set)
ridge_model = Ridge(alpha=0.1).fit(train_prepared, label)
ridge_model.score(train_prepared, label)
```

```python
# Elastic Net
num_attr = [column for column in train_set.columns if not train_set[column].dtype == 'object']
cat_attr = [column for column in train_set.columns if train_set[column].dtype == 'object']

num_processing = Pipeline([
    ('num', SimpleImputer(strategy='mean')),
    ('poly', PolynomialFeatures(degree=3, include_bias=False)),
    ('scale', StandardScaler()),
])

cat_processing = Pipeline([
    ('onehot', OneHotEncoder())
])

data_processing = ColumnTransformer([
    ('num_feature', num_processing, num_attr),
    ('cat_features', cat_processing, cat_attr), 
])

train_prepared = data_processing.fit_transform(train_set)
elastic_model = ElasticNet(max_iter=10000).fit(train_prepared, label)
elastic_model.score(train_prepared, label)
```

```python
# Svr
num_attr = [column for column in train_set.columns if not train_set[column].dtype == 'object']
cat_attr = [column for column in train_set.columns if train_set[column].dtype == 'object']

num_processing = Pipeline([
    ('num', SimpleImputer(strategy='mean')),
    ('scale', StandardScaler()),
])

cat_processing = Pipeline([
    ('onehot', OneHotEncoder())
])

data_processing = ColumnTransformer([
    ('num_feature', num_processing, num_attr),
    ('cat_features', cat_processing, cat_attr), 
])

train_prepared = data_processing.fit_transform(train_set)
svr_model = SVR().fit(train_prepared, label)
svr_model.score(train_prepared, label)
```

```python
# random forest 
num_attr = [column for column in train_set.columns if not train_set[column].dtype == 'object']
cat_attr = [column for column in train_set.columns if train_set[column].dtype == 'object']

num_processing = Pipeline([
    ('num', SimpleImputer(strategy='mean')),
    ('scale', StandardScaler()),
])

cat_processing = Pipeline([
    ('onehot', OneHotEncoder())
])

data_processing = ColumnTransformer([
    ('num_feature', num_processing, num_attr),
    ('cat_features', cat_processing, cat_attr), 
])

train_prepared = data_processing.fit_transform(train_set)
rf_model = RandomForestRegressor().fit(train_prepared, label)
rf_model.score(train_prepared, label)
```
